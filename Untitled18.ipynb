```python
# ==============================
# REINFORCEMENT LEARNING PROJECT
# ==============================

import numpy as np
import random

# ============================================
# PROJECT 1: NUMBER GUESSING RL AGENT
# ============================================

print("===== PROJECT 1: Number Guessing RL =====\n")

Q1 = np.zeros((11, 11))
episodes = 500
epsilon = 0.3
alpha = 0.1

# Training
for episode in range(episodes):
    target = random.randint(0, 10)

    for step in range(10):
        if random.random() < epsilon:
            action = random.randint(0,10)
        else:
            action = np.argmax(Q1[0])

        reward = 1 if action == target else -0.1
        Q1[0][action] += alpha * (reward - Q1[0][action])

        if action == target:
            break

print("Training Done!\n")

# Testing
correct = 0
for _ in range(10):
    target = random.randint(0,10)
    guess = np.argmax(Q1[0])
    print(f"Target: {target}, Guess: {guess}")

    if guess == target:
        correct += 1

print(f"\nAccuracy: {correct}/10\n")


# ============================================
# PROJECT 2: GRID WORLD RL AGENT
# ============================================

print("===== PROJECT 2: Grid World RL =====\n")

states = 5
actions = 4

Q2 = np.zeros((states, actions))

episodes = 500
alpha = 0.1
gamma = 0.9
epsilon = 0.3

goal_state = 4

# Training
for episode in range(episodes):
    state = 0

    while state != goal_state:
        if random.random() < epsilon:
            action = random.randint(0,3)
        else:
            action = np.argmax(Q2[state])

        next_state = min(state + 1, goal_state)
        reward = 1 if next_state == goal_state else -0.1

        Q2[state][action] += alpha * (reward + gamma * np.max(Q2[next_state]) - Q2[state][action])

        state = next_state

print("Training Complete!\n")
print("Final Q-Table:\n", Q2)
```
